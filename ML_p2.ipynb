```python
import pandas as pd
```

```python
input_file = 'Sales_Data.csv'
output_file = 'updated_file.csv'
```

```python
#import data from csv file
sales_data = pd.read_csv(input_file)
print("data imported sucessfully")
sales_data.head()
```

```python
sales_data = sales_data.drop(columns=['Region','Country','Order_Priority'])
```

```python
sales_data.head()
```

```python
sales_data = sales_data.drop(columns=['Sales_Channel','Order_Date','Ship_Date','Unit_MakingCost'])
sales_data.head()
```

```python
sales_data.isnull().sum()
```

```python
#handel null rows

sales_data = sales_data.dropna()
sales_data.isnull().sum()
```

```python
sales_data['total revenue'] = sales_data['Units_Sold'] * sales_data['Unit_SellingPrice']
sales_data.head()
```

```python
sales_data['sales tax'] = sales_data['total revenue'] * 0.10
sales_data['total sales'] = sales_data['total revenue'] + sales_data['sales tax']
sales_data.head()
```

```python
#export the uploaded dataframes to a new csv file
sales_data.to_csv(output_file, index=False)
print("data exported sucessfully")
```
